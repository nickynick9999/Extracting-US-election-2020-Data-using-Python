```python
import tweepy
import time
import re 
import csv 
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.feature_extraction.text import CountVectorizer
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize 
import string
import re
%matplotlib inline
```

```python
#Authentication key
consumer_key=""
consumer_secret=""
access_token=""
access_token_secret=""
```

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth,wait_on_rate_limit=True)
```

```python
keyword = ['Trump','Biden','Pence','Harris']
limit = 2000
try:
    # Creation of query method using parameters
    tweets = tweepy.Cursor(api.search,q=keyword).items(limit)

    # Pulling information from tweets iterable object
    tweets_list = [[tweet.created_at,tweet.user.name, tweet.text ] for tweet in tweets]

    # Creation of dataframe from tweets list
    # Add or remove columns as you remove tweet information
    tweets_df = pd.DataFrame(tweets_list)

except BaseException as e:
    print('failed on_status,',str(e))
    time.sleep(3)
```

```python
tweets_df.columns=['Time tweeted','Username','Tweet']
```

```python
string.punctuation
```

```python
def remove_punct(text):
    text="".join([char for char in text if char not in string.punctuation])
    text=re.sub('[0-9]+','',text)
    return text

tweets_df['Punct_Tweet']=tweets_df['Tweet'].apply(lambda x:remove_punct(x))
```

```python
def tokenization(text):
    text = re.split('\W+', text)
    return text

tweets_df['Tokenize_Tweet'] = tweets_df['Punct_Tweet'].apply(lambda x: tokenization(x.lower()))
```

```python
stopword = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(text):
    text = [word for word in text if word not in stopword]
    return text
    
tweets_df['nonStopWordsTweet'] = tweets_df['Tokenize_Tweet'].apply(lambda x: remove_stopwords(x))
tweets_df.head(10)
```

```python
ps = nltk.PorterStemmer()

def stemming(text):
    text = [ps.stem(word) for word in text]
    return text

tweets_df['StemmedTweets'] = tweets_df['nonStopWordsTweet'].apply(lambda x: stemming(x))
tweets_df.head()
```

```python
wn = nltk.WordNetLemmatizer()

def lemmatizer(text):
    text = [wn.lemmatize(word) for word in text]
    return text

tweets_df['LemmantizeTweets'] = tweets_df['nonStopWordsTweet'].apply(lambda x: lemmatizer(x))
tweets_df.head()
```

```python
tweets_df['Punct_Tweet']=tweets_df['Punct_Tweet'].astype(str)
```

```python
txt=tweets_df['Punct_Tweet'].to_json()
```

```python
stop_words = set(stopwords.words('english'))
tokenized=sent_tokenize(txt)
for i in tokenized: 
      
    # Word tokenizers is used to find the words  
    # and punctuation in a string 
    wordsList = nltk.word_tokenize(i) 
  
    # removing stop words from wordList 
    wordsList = [w for w in wordsList if not w in stop_words]  
  
    #  Using a Tagger. Which is part-of-speech  
    # tagger or POS-tagger.  
    tagged = nltk.pos_tag(wordsList) 
  
    print(tagged) 
```

```python
res=len(re.findall(r'\w+',txt))
print("The number of words in tweets are :"+str(res))
```
